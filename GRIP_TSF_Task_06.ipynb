# Prediction using Decision Tree Algorithm

**Step 1: Importing necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

**Step 2: Loading the IRIS Dataset**

```python
# Load the IRIS dataset
iris_data = load_iris()
X = iris_data.data
y = iris_data.target
print("Data import successfull!")
```

```python
# X = iris_data.iloc[:, :-1].values
# Y = iris_data.iloc[:, 1].values
```

**Step 3: Splitting data for training and testing**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

**Step 4: Create and Train the Decision Tree Classifier**

```python
# Create a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=50)

# Train the classifier using the training data
clf.fit(X_train, y_train)
```

**Step 5: Making Predictions**

```python
# Use the trained classifier to make predictions on the testing data
y_pred = clf.predict(X_test)
```

**Step 6: Evaluating the Classifier using sklearn library**

```python
# Evaluate the classifier's performance (e.g., accuracy)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
format_accuracy = "{:.2%}".format(accuracy)
print("Accuracy:", format_accuracy)
```

**Step 7: Visualizing the Decision Tree**

```python
# Visualize the Decision Tree graphically
plt.figure(figsize=(12, 8))
tree.plot_tree(clf, feature_names=iris_data.feature_names, class_names=iris_data.target_names, filled=True)
plt.show()
```
